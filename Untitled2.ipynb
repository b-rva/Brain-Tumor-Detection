Create dataframe from directory

```python
def create_df(path):
    class_paths, labels = [], []

    for label in os.listdir(path):
        class_folder = os.path.join(path, label)
        if os.path.isdir(class_folder):  # Ensure it's a folder
            for image in os.listdir(class_folder):
                class_paths.append(os.path.join(class_folder, image))
                labels.append(label)

    return pd.DataFrame({'Class Path': class_paths, 'Class': labels})
```

# **Data Preprocessing**

**Creating tf.data.Dataset Pipeline**

Constants

```python
BATCH_SIZE = 32
IMG_SIZE = (299, 299)
NUM_CLASSES = 4
```

Function to load and preprocess images

```python
def parse_image(filename, label):
    img = tf.io.read_file(filename)
    img = tf.image.decode_jpeg(img, channels=3)  # Decode JPEG images (change if needed)
    img = tf.image.resize(img, IMG_SIZE)  # Resize to match model input size
    img = img / 255.0  # Normalize pixel values
    label = tf.one_hot(label, depth=NUM_CLASSES)  # Convert to one-hot
    return img, label
```

Convert class labels to numerical format

```python
class_names = sorted(tr_df["Class"].unique())  # Get unique class labels
label_to_index = {name: idx for idx, name in enumerate(class_names)}  # Map labels to integers
```

Create a dataset

```python
def create_dataset(df, shuffle=True):
    file_paths = df["Class Path"].values
    labels = df["Class"].map(label_to_index).values  # Convert labels to numbers

    dataset = tf.data.Dataset.from_tensor_slices((file_paths, labels))
    dataset = dataset.map(parse_image, num_parallel_calls=tf.data.AUTOTUNE)  # Apply preprocessing
    if shuffle:
        dataset = dataset.shuffle(len(df))  # Shuffle dataset for training
    dataset = dataset.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)  # Optimize performance
    return dataset
```

Create train, validation, and test datasets

```python
train_ds = create_dataset(tr_df)
valid_ds = create_dataset(valid_df, shuffle=False)  # No shuffle for validation
test_ds = create_dataset(ts_df, shuffle=False)  # No shuffle for test data
```

# **Define model with transfer learning**

Define input shape

```python
IMG_SHAPE = (299, 299, 3)
```

Load Xception model (pretrained on ImageNet)

```python
base_model = Xception(weights="imagenet", include_top=False, input_shape=IMG_SHAPE)
```

Freeze base model layers

```python
base_model.trainable = False
```

Define the model architecture

```python
inputs = Input(shape=IMG_SHAPE)
x = base_model(inputs, training=False)  # Keep batch normalization layers in inference mode
x = GlobalAveragePooling2D()(x)  # Better than Flatten()
x = Dropout(0.3)(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.25)(x)
outputs = Dense(4, activation='softmax')(x)  # 4 classes
```

Create final model

```python
model = Model(inputs, outputs)
```

Compile the model

```python
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss="categorical_crossentropy",
    metrics=["accuracy", Precision(), Recall()]
)
```

# **Predict on a New Image**

```python
def predict(img_path, model, class_dict):
    class_names = ["glioma", "meningioma", "notumor", "pituitary"]

    # Load and preprocess image
    img = Image.open(img_path)
    resized_img = img.resize((299, 299))  # Resize to match input size
    img_array = np.asarray(resized_img) / 255.0  # Normalize
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension

    # Get model predictions
    predictions = model.predict(img_array)
    probs = list(predictions[0])

    # Display image and probability distribution
    plt.figure(figsize=(12, 12))
    plt.subplot(2, 1, 1)
    plt.imshow(resized_img)
    plt.axis("off")
    plt.title("Input Image")

    plt.subplot(2, 1, 2)
    bars = plt.barh(labels, probs, color='royalblue')
    plt.xlabel('Probability', fontsize=15)
    ax = plt.gca()
    ax.bar_label(bars, fmt='%.2f')
    plt.show()

    # Print predicted class
    predicted_label = labels[np.argmax(probs)]
    print(f"Predicted Class: {predicted_label} (Confidence: {max(probs)*100:.2f}%)")
```

```python
import matplotlib.pyplot as plt
from matplotlib.table import Table

# Data for the table
data = [
    ["Accuracy", "Fluctuates across epochs", "Consistently high"],
    ["Training Time", "Longer due to training from scratch", "Faster due to pre-trained weights"],
    ["Convergence", "Slower, requires more epochs", "Rapid convergence in fewer epochs"],
    ["Misclassifications", "Higher, struggles with complex cases", "Lower, robust in classification"],
    ["Computation", "Requires more resources", "Optimized for efficiency"]
]

# Column labels
columns = ["Feature", "Custom CNN", "Xception"]

# Create figure and axis
fig, ax = plt.subplots(figsize=(9, 4))
ax.set_xlim(0, 3)
ax.set_ylim(0, len(data) + 1)
ax.axis("off")

# Create table
table = Table(ax, bbox=[0, 0, 1, 1])

# Define column widths
col_widths = [0.3, 0.35, 0.35]

# Add headers with a background color
for i, col in enumerate(columns):
    table.add_cell(0, i, col_widths[i], 0.5, text=col, loc="center", facecolor="#DDDDDD")

# Add data with word wrapping
for row_idx, row in enumerate(data, start=1):
    for col_idx, cell_text in enumerate(row):
        table.add_cell(row_idx, col_idx, col_widths[col_idx], 0.5, text=cell_text, loc="center")

# Add the table to the plot
ax.add_table(table)

# Save as image
plt.savefig("model_comparison1.png", dpi=300, bbox_inches="tight")
plt.show()
```
